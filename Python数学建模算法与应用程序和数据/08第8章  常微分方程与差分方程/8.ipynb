```python
import sympy as sp
import numpy as np
```

```python
x = sp.var("x")
y = sp.Function("y")
eq = y(x).diff(x) + 2 * y(x) - 2 * x ** 2 + 2 * x
s = sp.dsolve(eq, ics={
    y(0): 1
})
print(s)
s = sp.simplify(s)
print(s)
s = sp.lambdify(x, s.args[1], "numpy")
s(1)
```

```python
x = sp.var("x")
y = sp.Function("y")
eq = y(x).diff(x, 2) - 2 * y(x).diff(x) + y(x) - sp.exp(x)
s = sp.dsolve(eq, ics={
    y(0): 1,
    y(x).diff(x).subs(x, 0): -1
})
s
```

```python
t = sp.var("t")
y = sp.Function("y")
u = sp.exp(-t) * sp.cos(t)
eq = y(t).diff(t, 4) + 10 * y(t).diff(t, 3) + 35 * y(t).diff(t, 2) + 50 * y(t).diff(t) + 24 * y(t) - u.diff(t, 2)
s = sp.dsolve(eq, ics={
    y(0): 0,
    y(t).diff(t).subs(t, 0): -1,
    y(t).diff(t, 2).subs(t, 0): 1,
    y(t).diff(t, 3).subs(t, 0): 1
})
print(s)
s = sp.expand(s)
s
```

```python
t = sp.var('t')
x1, x2, x3 = sp.var('x1:4', cls=sp.Function)  #定义3个符号函数
x = sp.Matrix([x1(t), x2(t), x3(t)])  #列向量
A = sp.Matrix([
    [3, -1, 1],
    [2, 0, -1],
    [1, -1, 2]
])
eq = x.diff(t) - A @ x
s = sp.dsolve(eq, ics={x1(0): 1, x2(0): 1, x3(0): 1})
s[0]
```

```python
#程序文件ex8_7.py
from scipy.integrate import odeint
import pylab as plt

dy = lambda y, x: -2 * y + 2 * x ** 2 + 2 * x  #自变量参数在函数参数最后
xx = np.linspace(0, 3, 31)
s = odeint(dy, 1, xx)
print('x={}\n对应的数值解y={}'.format(xx, s.flatten()))
plt.plot(xx, s, '*')
sp.var('x');
y = sp.Function('y')
eq = y(x).diff(x) + 2 * y(x) - 2 * x ** 2 - 2 * x
s2 = sp.dsolve(eq, ics={y(0): 1})
sx = sp.lambdify(x, s2.args[1], 'numpy')  #符号函数转匿名函数
plt.plot(xx, sx(xx))
plt.show()
```

```python
#程序文件ex8_9.py
from scipy.integrate import odeint
import numpy as np
import pylab as plt

np.random.seed(2)  #为了进行一致性比较，每次运行取相同随机数
sigma = 10
rho = 28
beta = 8 / 3
g = lambda f, t: [
    sigma * (f[1] - f[0]),
    rho * f[0] - f[1] - f[0] * f[2],
    f[0] * f[1] - beta * f[2]]  #定义微分方程组的右端项
s01 = np.random.rand(3)  #初始值
t0 = np.linspace(0, 50, 5000)
s1 = odeint(g, s01, t0)  #求数值解
s02 = s01 + 0.000001
s2 = odeint(g, s02, t0)  #初值变化后，再求数值解
```

```python
import plotly.express as px

px.scatter_3d(x=s1[:, 0], y=s1[:, 1], z=s1[:, 2])
```

```python
px.scatter(x=t0, y=s1[:, 0] - s2[:, 0])
```

```python
df = lambda f, t, w: [
    w / np.sqrt((10 + 20 * np.cos(t) - f[0]) ** 2 + (20 + 15 * np.sin(t) - f[1]) ** 2) * (10 + 20 * np.cos(t) - f[0]),
    w / np.sqrt((10 + 20 * np.cos(t) - f[0]) ** 2 + (20 + 15 * np.sin(t) - f[1]) ** 2) * (20 + 15 * np.sin(t) - f[1])]
d = lambda xy, t: np.sqrt(
    (xy[:, 0] - 10 - 20 * np.cos(t)) ** 2 +
    (xy[:, 1] - 20 - 15 * np.sin(t)) ** 2)  #计算距离函数
```

```python
t1 = np.linspace(0, 3.4068, 100)
s1 = odeint(df, [0, 0], t1, args=(20,))
d1 = d(s1, t1)
px.line(x=t1, y=d1)
```

```python
t2 = np.linspace(0, 50, 501)
s2 = odeint(df, [0, 0], t2, args=(5,))
d2 = d(s2, t2)
px.line(x=t2, y=[d2, d2 + 10], color_discrete_sequence=["blue", "red"])
```

```python
#程序文件ex8_13.py
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

a = pd.read_excel('data8_13.xlsx', header=None)
b = a.values
xd = b[1::2, :]
xd = xd[~np.isnan(xd)]  #提出有效数据
td = np.linspace(1790, 2000, 22)
x = lambda t, r, xm: xm / (1 + (xm / 3.9 - 1) * np.exp(-r * (t - 1790)))
bd = [(0, 200), (0.1, 1000)]  #约束两个参数的下界和上界
p1 = curve_fit(x, td[1:], xd[1:], bounds=bd)[0]  #拟合参数
print(p1)
print("2010年的预测值为：", round(x(2010, *p1), 4))


n = len(xd)
b1 = np.diff(xd) / 10 / xd[:-1]  #构造常数项列
a1 = np.vstack([np.ones(n - 1), -xd[:-1]]).T
p2 = np.linalg.pinv(a1) @ b1
r = p2[0]
xm = r / p2[1]
xh2 = x(2010, r, xm)
print('----------')
print(round(r, 4))
print(round(xm, 4))
print(round(xh2, 4))



b2 = np.diff(xd) / 10 / xd[1:]  #构造常数项列
a2 = np.vstack([np.ones(n - 1), -xd[1:]]).T
p3 = np.linalg.pinv(a2) @ b2
r = p3[0]
xm = r / p3[1]
xh3 = x(2010, r, xm)
print('----------')
print(round(r, 4))
print(round(xm, 4))
print(round(xh3, 4))
```
